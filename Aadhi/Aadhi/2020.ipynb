Plotting the graph for the Close Value of Apple Stock Price in Dollars($)

```python
plt.plot(data['Date'],data['Close/Last'],label="Close Value in $")
#plt.xlabel('Date')
#plt.ylabel("Market Price")
plt.title('Apple Close Price')
plt.legend(shadow=True)
plt.grid()
plt.show()
```

```python
plt.plot(data['Date'],data['Volume'],label="Volume in 100 Million")
#plt.xlabel('Date')
#plt.ylabel("Volume")
plt.title('Stock Trade Volume')
plt.legend(shadow=True)
plt.grid()
plt.show()
```
