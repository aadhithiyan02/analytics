```python
import pandas as pd
data=pd.read_excel("2020.xlsx")
data=data.set_index("Date")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.plot(data.index,data["Volume"])
plt.title("Volume Statistics")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
fig,axs=plt.subplots(4,1)
axs[0].plot(data.index,data["Open"])
axs[0].set_title("Open statistics")
axs[0].set_xlabel("Date")
axs[0].set_ylabel("Open")
axs[1].plot(data.index,data["High"])
axs[1].set_title("High statistics")
axs[1].set_xlabel("Date")
axs[1].set_ylabel("High")
axs[2].plot(data.index,data["Low"])
axs[2].set_title("Low statistics")
axs[2].set_xlabel("Date")
axs[2].set_ylabel("Low")
axs[3].plot(data.index,data["Close/Last"])
axs[3].set_title("Close statistics")
axs[3].set_xlabel("Date")
axs[3].set_ylabel("Close")
plt.subplots_adjust(top=7.0,bottom=5.0,wspace=1.0,hspace=1.0)

```
