```python
import pandas as pd
df=pd.read_excel("2021.xlsx")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df=df.set_index("Date")
df
```

```python
plt.plot(df.index,df['Volume'])
plt.title("Volume Statistics")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
fig,axs=plt.subplots(4,1)
axs[0].plot(df.index,df['Open'])
axs[0].set_title("Open Statistics")
axs[0].set_xlabel("Date")
axs[0].set_ylabel("Open")
axs[1].plot(df.index,df['High'])
axs[1].set_title("High Statistics")
axs[1].set_xlabel("Date")
axs[1].set_ylabel("High")
axs[2].plot(df.index,df['Low'])
axs[2].set_title("Low Statistics")
axs[2].set_xlabel("Date")
axs[2].set_ylabel("Low")
axs[3].plot(df.index,df['Close/Last'])
axs[3].set_title("Close Statistics")
axs[3].set_xlabel("Date")
axs[3].set_ylabel("Close/Last")
#plt.tight_layout()
#plt.subplot_tool()
plt.subplots_adjust(top=7.0,bottom=5.0,wspace=0.7,hspace=0.7)
plt.show()
```
