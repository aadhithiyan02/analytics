```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_frames = []
for sheet_name in ['2018', '2019', '2020', '2021', '2022']:
    data_frames.append(pd.read_excel("04_Netflix.xlsx", sheet_name=sheet_name))

netflix_total = pd.concat(data_frames)

print(netflix_total)
```

```python
netflix_total.plot(x = "Date", y = "Volume", figsize = (12,4))
plt.title("Volume(5 years)")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.grid()
```

```python
netflix_total.plot(x = "Date", y = "Close/Last" , figsize = (12,4))
plt.title("Close/Last(5years)")
plt.xlabel("Date")
plt.ylabel("Close/Last")
plt.grid()
```

```python
netflix = pd.read_excel("04_Netflix.xlsx")
```

```python
start_date = '2018-01-02'
end_date = '2018-12-31'
interval_2018 = netflix[(netflix['Date'] >= start_date) & (netflix['Date'] <= end_date)]
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(16,5))
ax1.plot(interval_2018['Date'],interval_2018['Close/Last'])
ax1.set_title("Close/Last", fontweight = 'bold')
ax1.set_xlabel("Date")
ax1.set_ylabel("Close/Last")


ax2.plot(interval_2018['Date'],interval_2018['Volume'])
ax2.set_title("Volume", fontweight = 'bold')
ax2.set_xlabel("Date")
ax2.set_ylabel("Volume")
plt.show()
 
```

### Based on the requirement we will be able to modify the years.

```python
years = [2018,2019,2020,2021,2022]
sheet_name = ['2018', '2019', '2020','2021','2022']

for i, year in enumerate(years):
    sheets = sheet_name[i % len(sheet_name)]
    data = pd.read_excel("04_Netflix.xlsx", sheet_name=sheets)
    
    start_date = pd.to_datetime(f"{year}-01-01")
    end_date = pd.to_datetime(f"{year}-12-31")
    
    interval_data = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

    ax1.plot(interval_data['Date'], interval_data['Close/Last'])
    ax1.set_title(f'Close/Last - {year}',fontweight = 'bold')
    ax1.set_xlabel("Date")
    ax1.set_ylabel("Close/Last")
    ax1.grid()

    ax2.plot(interval_data['Date'], interval_data['Volume'])
    ax2.set_xlabel("Date")
    ax2.set_ylabel("Volume")
    ax2.set_title(f'Volume - {year}',fontweight = 'bold')
    ax2.grid()
    
    fig.subplots_adjust(hspace=0.5)
    plt.show()
```
